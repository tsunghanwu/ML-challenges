## II. Data analysis on profitablity
* Calculate the profit and make it into a new column 
* Calculate the profit per shown and make it into a new column
* The top 5 ad gruops for profit are group 2, 31, 16, 14, 27
* The top 5 ad gruops for profit per shown are group 16, 2, 14, 31, 27

## III. Feature engineering
* Extract month, day, weekday features from date
* Create day_delta feature using the time difference between the first day and current day
* Extract the number of the ad group string as a feature

```python
first_day = df['date'].min()
df['month'] = df['date'].apply(lambda x: x.month)
df['day'] = df['date'].apply(lambda x: x.day)
df['day_delta'] = df['date'].apply(lambda x: (x - first_day).days)
df['weekday'] = df['date'].apply(lambda x: x.weekday())
```

```python
df.head()
```

```python
# Create a new columns with just group number

df['ad_group'] = df['ad'].apply(lambda x: int(x.split('_')[2]))
df.head()
```

## IV. Build ML model
### i. Visualize the trend for ad shown

```python
df['date'] = df['date'].apply(lambda x: x.date())
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(12,10))
sns.pointplot(x='date', y='shown', data=df, hue='ad', markers='+', linestyles='-')
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
```

### ii. Train ML models for different ad groups
* Take time related features as input features
* Target feature is 'shown'
* Use Random Forest Regressor as ML model
* Train the ML models for different ad groups

```python
# Select time related features for ML model

time_features = ['day_delta', 'weekday', 'month', 'day']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import median_absolute_error

# Train ML models for different groups

models = []
scores = []

for i in range(40):
    X = df[df['ad_group'] == (i+1)][time_features]
    y = df[df['ad_group'] == (i+1)]['shown']
    
    clf = RandomForestRegressor(n_estimators=100, random_state=100)
    clf.fit(X, y)
    y_pred = clf.predict(X)
    score = np.sqrt(median_absolute_error(y, y_pred))
    models.append(clf)
    scores.append(score)
```

## V. Evaluate ML models
### i. Evaluate the training results
* Visualize the predictions of the training data and actual data
* The training MAEs for the ad groups are mostly around 10 - 40

```python
# Create a function to plot the stack graph with actual shown curve and predicted shown curve
def plot_prediction(group_id, df):
    X = df[df['ad_group'] == (group_id)][time_features]
    y = df[df['ad_group'] == (group_id)]['shown']
    clf = models[group_id-1]
    y_pred = clf.predict(X)
    
    plt.plot(X['day_delta'], y, '-o', label='Actual', markersize=6)
    plt.plot(X['day_delta'], y_pred, '-o', label='Prediction', markersize=6)
    plt.xlabel('Day delta')
    plt.ylabel('Shown')
    plt.legend()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(14,24))
for i in range(40):
    plt.subplot(10, 4, i+1)
    plot_prediction(i+1, df)
    plt.title('ad_group = {}'.format(i+1))
plt.tight_layout()
```

```python
# Visualize the training MAE for different groups

df_scores = pd.DataFrame({'ad_group': np.arange(1,41), 'MAE': scores})

plt.style.use('seaborn')
sns.catplot(x='ad_group', y='MAE', data=df_scores, kind='bar', aspect=2, color='tab:blue')
```

### ii. Generate future predictions

```python
from datetime import datetime
from datetime import date
from datetime import timedelta

target_date = datetime(year=2015, month=12, day=15)
target_date - first_day, target_date.weekday()
```

```python
delta = target_date - first_day
date = []
for day in range(0, delta.days+1):
    date.append(first_day + timedelta(days=day))
```

```python
# Create a new datafrom with future dates

X_1215 = pd.DataFrame(columns=time_features)
X_1215['date'] = date

X_1215['month'] = X_1215['date'].apply(lambda x: x.month)
X_1215['day'] = X_1215['date'].apply(lambda x: x.day)
X_1215['day_delta'] = X_1215['date'].apply(lambda x: (x - first_day).days)
X_1215['weekday'] = X_1215['date'].apply(lambda x: x.weekday())

X_1215 = X_1215.drop('date', axis=1)
X_1215.head()
```

### iii. Evaluate the future predictions
* Most of the future prediction curves show good consistency with original curves

```python
# Create a function to plot the stack graph with actual shown curve and predicted shown curve, including furture predictions

def plot_prediction_1215(group_id, df, X_new):
    X = df[df['ad_group'] == (group_id)][time_features]
    y = df[df['ad_group'] == (group_id)]['shown']
    clf = models[group_id-1]
    y_pred_new = clf.predict(X_new)
    
    plt.plot(X['day_delta'], y, '-o', label='Actual')
    plt.plot(X_new['day_delta'], y_pred_new, '-o', label='Prediction')
    plt.xlabel('Day delta')
    plt.ylabel('Shown')
    plt.legend()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(14,24))
for i in range(40):
    plt.subplot(10, 4, i+1)
    plot_prediction_1215(i+1, df, X_1215)
    plt.title('ad_group = {}'.format(i+1))
plt.tight_layout()
```

```python
# Create a dataframe with future predictions

df_1215_pred = pd.DataFrame(columns=['date'])
df_1215_pred['date'] = date

for i in range(40):
    clf = models[i-1]
    y_pred_1215 = clf.predict(X_1215)
    df_1215_pred['group {}'.format(i+1)] = y_pred_1215 
```

## VI. Cluster by avg_cost_per_click trend
### i. Visualize the trends

```python
plt.style.use('seaborn')
plt.figure(figsize=(14,24))
for i in range(40):
    plt.subplot(10, 4, i+1)
    sns.regplot(x='day_delta', y='avg_cost_per_click', data=df[df['ad_group'] == (i+1)], robust=True)
    plt.title('ad_group = {}'.format(i+1))
plt.tight_layout()
```

### ii. Fit the regression models

```python
from sklearn.linear_model import HuberRegressor

# Create a function to generate slope coeff from the Huber regression model

def find_trend(X, y):
    clf = HuberRegressor()
    clf.fit(np.array(X).reshape(len(X), 1), np.array(y).reshape(len(y), ))
    return clf.coef_[0]
```

### iii. Extract the slope coefficients
* Determine the decision boundaries are +/- 0.001

```python
# Collect the slope coeffs from different groups

trends = []

for i in range(40):
    X = df[df['ad_group'] == (i+1)]['day_delta']
    y = df[df['ad_group'] == (i+1)]['avg_cost_per_click']
    trend = find_trend(X, y)
    trends.append(trend)
```

```python
# Find the 25%, 50% and 75% percentile of the slope coeffs

np.percentile(trends, 25), np.percentile(trends, 50), np.percentile(trends, 75)
```

```python
# Choose +/- 0.001 as decision boundaries

plt.hist(trends, bins=40)
plt.axvline(-0.001, ls='--')
plt.axvline(0.001, ls='--')
plt.xlabel('Trend');
```

### iv. Find the trend-up / no-trend / trend-down groups
* Trend up: 1, 3, 20, 32, 37, 40
* Trend down: 4, 6, 7, 8, 19, 24, 28, 29, 33, 36, 38
* No trend: Rest of the groups

```python
df_trend = pd.DataFrame(columns=['group', 'trend'])
df_trend['trend'] = trends
df_trend['group'] = np.arange(1, 41)
```

```python
# Trend up groups

df_trend[df_trend['trend'] >= 0.001]
```

```python
# No trend groups

df_trend[(df_trend['trend'] < 0.001) & (df_trend['trend'] > -0.001)]
```

```python
# Trend down groups

df_trend[df_trend['trend'] <= -0.001]
```
